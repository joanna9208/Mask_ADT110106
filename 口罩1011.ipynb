```python
# prgl

import csv
import requests
import pandas as pd

prefix = ''

url = "https://data.nhi.gov.tw/resource/mask/maskdata.csv"

prefix_list = []
row_count = 0

with requests.Session() as s:
  download = requests.get(url)
  decoded_content = download.content.decode('utf-8')
  cr = csv.reader(decoded_content.splitlines(), delimiter=",")

  original_list = list(cr)
  #print(original_list)
  print(' Original_list len = %d' % (len(original_list)-1))

  # prefix 
  # ----------------------------------
  for row in original_list:
    row_count = row_count + 1  #1
    if row_count != 1:
      if prefix == '':
        prefix_list.append(row)
      elif row[2].startswith(prefix):     #prefixprefix_list
        prefix_list.append(row)
    print('prefix_list len= %d' % len(prefix_list)) #

def myFunc(row):
  return int(row[4])   #row[4]


prefix_list.sort(key=myFunc, reverse=True)   #

prefix_list.insert(0, original_list[0])  #prefix_list

mask_df = pd.DataFrame(prefix_list) #prefix_listpdDataFramemask_df
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(mask_df) #mask_df

#mask_df = pd.DataFrame(original_list)
#mask_df.head()
```
